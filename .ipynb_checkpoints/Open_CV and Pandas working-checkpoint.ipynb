# Use of CV2 and Image processing in pyplot

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

the cv2 takes the color as BGR and not RGB thats y we have to change the color of the cv2 by using the inbuilt function

```python
img = cv2.imread("car.jpg")
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # here we have changed the colored pic to grey scale
```

```python
print(img.shape)
```

```python
plt.imshow(img)
plt.axis("off") # this is used to not showing the axis on the picture
plt.show()
```

```python
plt.imshow(rgb_img)
plt.show()
# this is the real picture in the pyplot method 
```

```python
crop_img = rgb_img[300:600,200:1000,:] # here the funciton takes arguments as [row:column:channels or features]
plt.imshow(crop_img)
plt.show()
# here we have croped the image 
```

```python
crop_img.shape
```

```python
plt.imshow(grey_img)
plt.show()
```

```python
total_pixal = crop_img.shape[0]*crop_img.shape[1] # to calculte the total pixel in the crop image
print(total_pixal)
```

```python
# resizing the image that we have cropped earlier
c = cv2.resize(crop_img,(100,100))
plt.imshow(c)
# we resize the crop image so that it takes less space and is easy to transfer this reduce its size and resolution
```

# Creating a DataFrame

```python
# Creating a dataset of pokemons
pokemon = { 
    'pokemon_name':np.array(["pikachu","balbasur","raichu","onex","firefox"]),
    'pokemon_height':np.random.randint(10,100,5),
    'pokemon_weight':np.random.randint(100,500,5),
    'pokemon_power':np.random.randint(200,1000,5)
}
```

```python
pokemon
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(pokemon)
print(df)
```

```python
df.head()
```

```python
df.columns
```

### Create a CSV file(Comma Seperated Valued File)

```python
df.to_csv("pokemon.csv") # this is the file name that will we ceated as the csv format
```
