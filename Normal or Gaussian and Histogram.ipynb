#  Noraml Distribution and Histogram for the univerate or simple distribution 

```python
# first we generate the noramal distribution data
import numpy as np
import matplotlib.pyplot as plt
mean = 0
sigma = 1
# above are the values for standard normal distribution
mean = 5
sigma =1
vals = np.random.randn(100) # here we are generating the data values for the 
```

```python
print(vals.shape)
```

```python
vals = (sigma*vals)+mean
```

```python
print(vals.shape)
```

```python
plt.hist(vals,50,color = "green") # this is the histogram plotting and here 50 is the width of the bar in the hist
plt.show()
```

* in the above histogram we can see that the data spread is more around the mean that is 0 and the spread of the graph is as per the sigma
* here each bar is shoeing how much data points are there in that area or inside the bar... consider it as a bucket of data points.
* here the mean is changed so does the graph new data points they are max at the mean points

```python
data = np.round(vals)
z = np.unique(data,return_counts=True)
# this tells that how many points are there in the data set and what is the frequency of each point
print(z)
```

```python
X = vals
y = np.zeros(X.shape)
plt.scatter(X,y)
plt.show()
```

* in the above we can see that the maximun points are near about the mean which is 5 here
* since these are the normal distribution spread is more near the mean

# Multivariate Normal distribution

 this is the normal distrubution with multi dimennsional ... this means that we have more than one features. But here we are finding the probability distribution and this is 3D graph... but we plot the points as the points on the surface .

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])
dist = np.random.multivariate_normal(mean,cov,500)

plt.scatter(dist[:,0],dist[:,1])
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])
# this is the first data set meana
dist = np.random.multivariate_normal(mean,cov,500)

mean2 = np.array([5.0,5.0])
cov2 = np.array([[1,0.5],[0.5,1]])
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist2.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
