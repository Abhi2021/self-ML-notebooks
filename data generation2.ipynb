# Classification data generation

we are going to use multivariate normal distribution for the below example. this is used bcoz we have more than one feature in the training data set. if we had only one feature than we can use normal distribution for the training data.

* here we are using two features as size and sweetness . 
* For both the fruits as apple and orange .  
* Then we will integrate both the data into one matrix and call it as X 
* Then Y will be calculated by the help of them and its matrix will be created.
* then we will show the data on the graph as two matrices in one and that will be shown as different class 

* here we also have the concept of mean and co-variance
* to generate the data we need the covarance and mean
* matrix of [[&xx, &xy],[&xy, &yy]] will be created this will show how the data is depended on each other.
* matrix 1 of x1 and x2 for apples is created and same for the oranges as matrix 2
* then we will combine them as one and generate Y
* then finally we plot the graph as scatter

```python
u1 = np.array([2,4])
cov1 = np.array([[0,1],[1,0]])

u2 = np.array([-1,-3])
cov2 = np.array([[1,0],[0,2]])
```

```python
X1 = np.random.multivariate_normal(u1, cov1, 300) # mean , covariance , number of points
X2 = np.random.multivariate_normal(u2, cov2, 400) # mean , covariance , number of points
```

```python
plt.scatter(X1[:,0],X1[:,1], label= "apple")
plt.scatter(X2[:,0],X2[:,1], label= "oranges")
plt.legend()
plt.xlabel("x1 size")
plt.ylabel("x2 sweetness")
plt.show()
```

```python
y= np.zeros(700) # matrix with 700 zeros as in the matrix, size if dicided by the compiler
```

```python
y[:300]=1
print(y.shape)
print(y)
```

```python
X = np.vstack((X1,X2))
```
