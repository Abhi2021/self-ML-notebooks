## 2. Data generation for classification

* these are descrete point data 
* 1.let us consider we have two classes apples and oranges with two features like size and sweetness. x1 = size and x2 = sweetness and some y values for them .
* 2.we will tell here weather the fruit is apple for oranges when we divide the data into test and training
 * NORMAL DISTRIBUTION is used when we have only one feature and MULTIVARIATE is used when we hae more than one feature
 * we need mean and covarance 
 * 

```python
u1 = np.array([2,4])
cov1 = np.array([[1,0],[0,1]])

u2 = np.array([-1,-3])
cov2 = np.array([[1,0],[0,2]])
```

```python
X1 = np.random.multivariate_normal(u1,cov1,300)
X2 = np.random.multivariate_normal(u2,cov2,300)
print(X1.shape)
print(X2.shape)
```

```python
plt.scatter(X1[:,0],X1[:,1], color= "red", label = "apples") # we are using X1[:,0] goes to the x axis and X1[:,1] goes to the y axis
plt.scatter(X2[:,0],X2[:,1],color = "green", marker = "^", label = "oranges") 
plt.legend()
plt.xlabel("size")
plt.ylabel("fruitness")
plt.show()
```

we have 300 data points with 2 features ... this is one cluester
,we want another cluster

* if we decrease the &xy but > 0 we will find the data has tilted towards the x axis by y =x(that means a positve relation) 
* but if we make it negative the data is tilted to y= -x line( this means the negative relation), if we dec the size of fruit the sweetness increases

```python
X = np.vstack((X1,X2))
print(X.shape)
```

```python
Y = np.zeros(600)
Y[:300] =1 
print(Y.shape)
```

```python
print(Y)
```

```python
plt.scatter(X[:,0],X[:,1], c =Y)
plt.show()
```

### make_regression

```python
from sklearn.datasets import make_regression
```
